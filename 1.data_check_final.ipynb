## 1. Libraries einlesen

````python
import pandas as pd
import numpy as np
````

## 2. Datensätze einlesen und Kopien erstellen

````python
df_brands_org = pd.read_csv("Daten/brands.csv")
df_olines_org = pd.read_csv("Daten/orderlines.csv")
df_orders_org = pd.read_csv("Daten/orders.csv")
df_products_org = pd.read_csv("Daten/products.csv")

df_brands = df_brands_org.copy()
df_olines = df_olines_org.copy()
df_orders = df_orders_org.copy()
df_products = df_products_org.copy()
````

### 2.1 Datentypen herausfinden

Einfache Methode:

````python
display(df_brands.info())
display(df_olines.info())
display(df_orders.info())
display(df_products.info())
````

### Darsellung in Tabelle
+ fehlende Werte werden mit X ersetzt
+ Ausgabe in CSV und Txt Format

````python
dt_vergleich = pd.concat([
    df_brands.dtypes.rename('Brands'),
    df_olines.dtypes.rename('Order Lines'),
    df_orders.dtypes.rename('Orders'),
    df_products.dtypes.rename('Products')
], axis=1)

display(dt_vergleich.fillna(' '))

# Die Tabelle mit 'X' füllen und als CSV exportieren
dt_vergleich.fillna('X').to_csv('datentypen_uebersicht.csv', index=True, sep=';', encoding='utf-8')
# Die Tabelle als formatierten Text in eine .txt-Datei schreiben
with open('datentypen_uebersicht.txt', 'w', encoding='utf-8') as f:
    f.write(dt_vergleich.fillna(' ').to_string())
````

### 2.2 Datentype anpassen
Anpassen der Zeile für den passenden Dtype

**2.2.1 Datumsspalten konvertieren**

````python
df_olines['date'] = pd.to_datetime(df_olines['date'], errors='coerce')
df_orders['created_date'] = pd.to_datetime(df_orders['created_date'], errors='coerce')
````

**2.2.2 Preise in Fließkommazahlen umwandeln**

````python

df_olines['unit_price'] = pd.to_numeric(df_olines['unit_price'], errors='coerce')
df_products['price'] = pd.to_numeric(df_products['price'], errors='coerce')
df_products['promo_price'] = pd.to_numeric(df_products['promo_price'], errors='coerce')
````

## 5. Logikfehler beheben
Promo-Preis darf nicht größer sein als der reguläre Preis.

````python
# Zuerst Zeilen löschen, die bei der Typ-Umwandlung kaputtgegangen sind (unsaubere Werte)
df_products = df_products.dropna(subset=['price', 'promo_price'])

# Filter anwenden: Wir behalten nur Zeilen, bei denen Promo-Preis <= Normalpreis ist
df_products = df_products[df_products['promo_price'] <= df_products['price']]
````

## 6. Fehlende Werte behandeln
Letzte verbleibende NaNs aus Orders (total_paid) und Products (desc, type) löschen.

````python
# Orderlines: Alle Zeilen löschen, bei denen der unit_price fehlt (NaN ist)
df_olines = df_olines.dropna(subset=['unit_price'])

# Products: Alle Zeilen löschen, bei denen price oder promo_price fehlt (NaN ist)
df_products = df_products.dropna(subset=['price', 'promo_price'])
df_products.drop(columns=['promo_price'], inplace=True)

# Products: Fehlende Werte in den Text-Spalten 'desc' und 'type' löschen
df_products = df_products.dropna(subset=['desc', 'type'])
````

## 7. Qualitätskontrolle (Check)
Prüfen, ob alle Tabellen sauber sind (0 Duplikate, 0 fehlende Werte).

````python
dataframes = {
    "Brands": df_brands,
    "Orderlines": df_olines,
    "Orders": df_orders,
    "Products": df_products
}

for name, df in dataframes.items():
    print(f"=== {name} ===")
    print(f"Zeilen / Spalten: {df.shape}")
    print(f"Duplikate: {df.duplicated().sum()}")
    fehlende = df.isna().sum().sum()
    print(f"Fehlende Werte (gesamt): {fehlende}")
    print("-" * 40)
````

## 8. Export
Saubere Datensätze abspeichern.

````python
df_brands.to_csv("Daten/clean/brands_clean.csv", index=False)
df_olines.to_csv("Daten/clean/orderlines_clean.csv", index=False)
df_orders.to_csv("Daten/clean/orders_clean.csv", index=False)
df_products.to_csv("Daten/clean/products_clean.csv", index=False)
print("Export-Code ist auskommentiert, zur Nutzung Rauten entfernen.")
````

## Auswertung & Kritik

Dieses Skript ist genau der Übeltäter, der für den massiven Umsatzverlust (von 15,7 Mio. € auf ca. 7-8 Mio. €) verantwortlich ist!

Warum? Beim Schritt 2.2.2 Preise in Fließkommazahlen umwandeln nutzt das Skript den Befehl pd.to_numeric(..., errors='coerce'). Das bedeutet: Alles, was kein perfekter Punkt ist (wie die Preise 1.137.99 mit zwei Punkten), wird sofort in NaN (Not a Number) verwandelt. Beim Schritt 6. Fehlende Werte behandeln löscht das Skript dann eiskalt alles, was ein NaN enthält.

Dadurch passiert Folgendes (wie man in Zelle 13 sieht):

Die products-Tabelle schrumpft auf mickrige 1.252 Produkte zusammen (vorher waren es über 10.000 einzigartige Produkte)! Fast 90% des Sortiments wurden gelöscht.
Eure saubere products_clean.csv hat also fast keine Produkte mehr, was den anschließenden Merge mit den Bestellungen ruiniert.
